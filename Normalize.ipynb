```python
input = '''
un ejemplo : 27/05/2008 : 42.2$ 
'''

#replaces
repl = {
 #ADD HERE
 '€': 'euros',
 '$': 'dolares', 
 '°': 'grados', 
 '#': 'hashtag', 
 '@': 'arobase', 
 '&': 'and'
}


#replace '<num>.<num>' by '<num> punto <num>'
previous_input_length = len(input)
temp_input = input
while(True):
    temp_input = re.sub(r'(.*\d)\.(\d.*)', r'\1 punto \2',   temp_input)
    len_temp_input = len(temp_input)
    if len_temp_input == previous_input_length:
        break
    previous_input_length = len_temp_input 
input = temp_input

#replace digits to words
input = input.replace('.', ' . ').replace('\'', ' \' ').replace('/', ' / ').replace(',', ' , ').replace('$', ' $ ').replace('€', ' € ').replace('£', ' £ ').replace('¥', ' ¥ ').replace('°', ' ° ')
l = [num2words(s, lang='es') if s.isdigit() else s for s in input.split() ]
output = ' '.join(l)
output = output.replace(' . ', '.').replace(' \' ', '\'').replace(' / ', '/').replace(' , ', ',').replace(' $ ', '$').replace(' € ', '€').replace(' £ ', '£').replace(' ¥ ', '¥').replace('°', ' ° ')

#replace replaces
l = [repl[s] if s in repl else s for s in output.split() ]
output = ' '.join(l)

print(output)
```
